# built simple regression model of sqft_living to price

```python
import numpy as np
training_set,test_set=np.split(df,[int(.8*len(df))])
training_set
```

```python
test_set
```

# Build the regression model

```python
from sklearn import linear_model
reg= linear_model.LinearRegression()

x=training_set.sqft_living
x=np.array(x)
x.reshape(-1,1)
y=training_set.price
y=np.array(y)
y.reshape(-1,1)

reg.fit(x,y)
```
